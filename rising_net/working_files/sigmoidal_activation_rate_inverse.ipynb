```python
import numpy as np
```

```python
beta = 20
sigma = 0.0
x = np.arange(-1.0, 1.0, 0.01)

S = 1.0/(1+np.exp(-beta*(x-sigma)))
```

```python
from matplotlib import pyplot
```

```python
pyplot.plot(x, S)
```

```python
xS = (sigma - np.log(1/S - 1)/beta)
```

```python
pyplot.plot(S, xS)
```
